```python
import json
import copy

qcopy = copy.deepcopy("v2_OpenEnded_mscoco_train2014_questions.json")
acopy = copy.deepcopy("v2_mscoco_train2014_annotations.json")
q = json.load(open(qcopy))
a = json.load(open(acopy))

```

```python
print(q.keys())
print(a.keys())
```

```python
qs = q['questions'][0]
ass = a['annotations'][0]
print(qs)
print(ass)
print(len(q['questions']))
print(len(a['annotations']))
```

```python
q_types = []
for ann in a['annotations']:
    if ann['question_type'] in q_types:
        continue
    q_types.append(ann['question_type'])
```

```python
print(q_types)
```

```python
color_ids = []
q_types = []
color_as = []
dict_idx = []
for idx, ann in enumerate(a['annotations']):
    if "color" not in ann['question_type']:
        continue
    if ann['question_type'] not in q_types:
        q_types.append(ann['question_type'])
    color_ids.append(ann['question_id'])
    color_as.append(ann)
    dict_idx.append(idx)
```
